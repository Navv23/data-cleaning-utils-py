```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/ign-games/ign.csv", index_col=[0])
```

```python
df.head()
```

```python
df = df.drop(['url'], axis=1)
```

```python
df.isnull().sum()
```

```python
df['genre'].value_counts()
```

```python
df = df.dropna(how='any')
```

```python
df['game_title'] = df['title'].str.upper()
```

```python
avg_genre_rating = df.groupby('genre')[['score']].mean().sort_values('score', ascending=False)
```

```python
avg_genre_rating
```

```python
df['date'] = df['release_year'].astype(str) + "-" + df['release_month'].astype(str) + "-" + df['release_day'].astype(str)
```

```python
df['date'] = pd.to_datetime(df['date'])
```
